a. Add columns name for dataframe with information such as: (ID,Name,Age,Weight,m0006,m0612,m1218,f006,f0612,f1218)

```python
data=pd.read_csv('https://raw.githubusercontent.com/huynhhoc/DataAnalystDeepLearning/main/Data/patient_heart_rate.csv',names=['ID','Name','Age','Weight','m0006','m0612','m1218','f006','f0612','f1218'])
data.head(6)
```

Multi-variables in one columns

```python
data[['First_Name','Last_Name']]=data['Name'].str.split(' ',expand=True)
```

```python
data.head(10)
```

del data['Name']

```python
data.head(10)
```

Inconsitence values. Convert "lbs" to "kg" with columns name "Weight".

```python
weight=data['Weight']
for i in range(len(weight)):
  x=str(weight[i])
  if "lbs" in x[-3:]:
    x=x[:-3:]
    y=int(float(x)/2.2)
    y=str(y)+'kgs'
    weight[i]=y
```

```python
data.head(10)
```

Drop An empty row[8,9] in the data

```python
data.dropna(how='all',inplace=True)
```

```python
data
```

5.Duplicate records in the data

In data, there is repeat in 6 and 9 row.

```python
data=data.drop_duplicates(subset=['Age','Weight','First_Name','Last_Name'])
data
```

Inconsitence characters (error ASCII-based format)
Error ASCII-based format in the Fisrt_Name and Last_Name columns in the data.

```python
data.First_Name.replace({r'[^\x00-\x7F]+':''},regex=True,inplace=True)
data.Last_Name.replace({r'[^\x00-\x7F]+':''},regex=True,inplace=True)
data
```

a. Melt the Sex + time range columns in single column

```python
data=pd.melt(data,id_vars=['ID', 'Age', 'Weight', 'First_Name', 'Last_Name'],value_name='PulseRate',var_name='sex_and_time').sort_values(['ID','Age','Weight','First_Name','Last_Name'])
data.head(10)
```

```python
tmp_data=data['sex_and_time'].str.extract('(\D)(\d+)(\d{2})',expand=True)
```

```python
tmp_data.columns=['Sex','Hour_Lower','Hour_Upper']
```

```python
tmp_data['Time']=tmp_data['Hour_Lower']+'-'+ tmp_data['Hour_Upper']
```

```python
#Merge "data and tmp_data":
data=pd.concat([data,tmp_data],axis=1)
```

```python
data.head(10)
```

e.Drop "sex_and_time,Hour_Lower,Hour_Upper" columns and row to data simple

```python
data=data.drop(['sex_and_time','Hour_Lower','Hour_Upper'],axis=1)
data=data.dropna()
```
